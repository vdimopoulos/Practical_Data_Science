```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("greeklish_translations.csv")
df
```

```python
df.columns
```

```python
df.drop(axis=1,columns=['Unnamed: 0'],inplace=True)
df
```

```python
import Levenshtein

def calculate_edit_distance(str1, str2):
    return Levenshtein.distance(str1, str2)
```

```python
edit_distance_matrix = np.empty((126,15))
```

```python
for i in range(126):
    l = 0
    for j in range(5):
        for k in range(j+1,6):
            edit_distance_matrix[i,l] = calculate_edit_distance(df.iloc[i,j],df.iloc[i,k])
            l += 1
edit_distance_matrix
```

```python
medians = np.median(edit_distance_matrix, axis=1)
lower_quartiles = np.percentile(edit_distance_matrix, 25, axis=1)
upper_quartiles = np.percentile(edit_distance_matrix, 75, axis=1)
iqr = upper_quartiles - lower_quartiles
sorted_indices = np.argsort(medians)[::-1]
sorted_matrix = edit_distance_matrix[sorted_indices]
sorted_iqr = iqr[sorted_indices]
fig, ax = plt.subplots(figsize=(8, 10))
ax.boxplot(sorted_matrix.T, vert=True,showfliers=False)  # Transpose for correct orientation
ax.set_xticks([])
ax.set_ylabel('Interquartile Range (IQR) of Edit Distance per Post')
ax.set_xlabel('Sorted Posts (by Median Edit Distance)')
ax.set_title("Plot of Posts' Edit Distance IQR")

plt.show()
```
